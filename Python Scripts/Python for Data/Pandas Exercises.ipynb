### Task: Dealing with the Titanic dataset

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
cleaned_data = titanic.copy()
```

```python
cleaned_data.drop_duplicates()
```

```python
missing_data = cleaned_data.isnull().sum()
missing_data
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(cleaned_data.isnull(), cbar=False)
```

```python
cleaned_data.drop(columns=['deck'], inplace=True)
```

```python
cleaned_data.age = cleaned_data.age.fillna(cleaned_data.age.mean())
cleaned_data = cleaned_data.ffill()
```

```python
before = titanic.size
print(before)
after = cleaned_data.size
print(after)
difference = before - after
print(f"Proportion of data removed: {round((difference / before) *100, 2)}%")

cleaned_data.isnull().sum()
```

```python
# Finding the average fare and average age for each class
avg_fare = cleaned_data.groupby(by='class').fare.mean()
avg_age = cleaned_data.groupby(by='class').age.mean()
avg_fare
```

```python
# Showing a count of female vs male survivors
cleaned_data[cleaned_data.survived == True].groupby(by='sex').sex.count()
```

```python
# Adding a new column showing the fare as a percentage of total
cleaned_data['fare as percentage'] = cleaned_data.fare / cleaned_data.fare.sum() * 100
cleaned_data.head()
```
